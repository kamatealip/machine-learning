# Gradient Descent on Multiple Linear Regression

This notebook shows how to train a multiple linear regression model using gradient descent for two input features.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Training data with two features
# y = 4 + 2 * x1 + 3 * x2
X = np.array([
    [1, 2],
    [2, 1],
    [3, 4],
    [4, 3],
    [5, 5],
    [6, 7]
], dtype=float)

y = np.array([12, 11, 22, 21, 29, 37], dtype=float)

print('Shape of X:', X.shape)
print('Shape of y:', y.shape)
print('\nX:\n', X)
print('\ny:\n', y)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, color='teal', s=60)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
ax.set_title('Multiple Linear Regression Data')
plt.show()
```

```python
def compute_cost(X, y, weights, bias):
    predictions = np.dot(X, weights) + bias
    return np.mean((predictions - y) ** 2)
```

```python
def gradient_descent(X, y, learning_rate=0.01, epochs=5000):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0.0
    cost_history = []

    for _ in range(epochs):
        predictions = np.dot(X, weights) + bias
        errors = predictions - y

        dw = (2 / n_samples) * np.dot(X.T, errors)
        db = (2 / n_samples) * np.sum(errors)

        weights = weights - learning_rate * dw
        bias = bias - learning_rate * db

        cost_history.append(compute_cost(X, y, weights, bias))

    return weights, bias, cost_history
```

```python
weights, bias, cost_history = gradient_descent(X, y, learning_rate=0.01, epochs=5000)

print('Weights:', np.round(weights, 4))
print('Bias:', round(bias, 4))
print('Final Cost:', round(cost_history[-1], 6))
```

```python
y_pred = np.dot(X, weights) + bias

for actual, predicted in zip(y, y_pred):
    print(f'Actual: {actual:.1f}, Predicted: {predicted:.2f}')
```

```python
plt.plot(cost_history, color='darkgreen')
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Cost Reduction During Training')
plt.show()
```

```python
plt.scatter(y, y_pred, color='crimson')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='black')
plt.show()
```

```python
new_data_point = np.array([7, 7], dtype=float)
predicted_value = np.dot(new_data_point, weights) + bias

print(f'Prediction for {new_data_point}: {predicted_value:.2f}')
```
