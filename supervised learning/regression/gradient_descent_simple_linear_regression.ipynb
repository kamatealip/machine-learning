# Gradient Descent on Simple Linear Regression

This notebook shows how to fit a straight line `y = mx + b` using gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple training data
X = np.array([1, 2, 3, 4, 5, 6], dtype=float)
y = np.array([3, 5, 7, 9, 11, 13], dtype=float)

plt.scatter(X, y, color='steelblue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Training Data')
plt.show()
```

```python
def compute_cost(X, y, m, b):
    predictions = m * X + b
    return np.mean((predictions - y) ** 2)
```

```python
def gradient_descent(X, y, learning_rate=0.01, epochs=1000):
    m = 0.0
    b = 0.0
    n = len(X)
    cost_history = []

    for _ in range(epochs):
        predictions = m * X + b

        dm = (-2 / n) * np.sum(X * (y - predictions))
        db = (-2 / n) * np.sum(y - predictions)

        m = m - learning_rate * dm
        b = b - learning_rate * db

        cost_history.append(compute_cost(X, y, m, b))

    return m, b, cost_history
```

```python
m, b, cost_history = gradient_descent(X, y, learning_rate=0.01, epochs=2000)

print('Slope (m):', round(m, 4))
print('Intercept (b):', round(b, 4))
print('Final Cost:', round(cost_history[-1], 6))
```

```python
y_pred = m * X + b

plt.scatter(X, y, color='steelblue', label='Actual Data')
plt.plot(X, y_pred, color='crimson', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Best Fit Line Using Gradient Descent')
plt.legend()
plt.show()
```

```python
plt.plot(cost_history, color='darkgreen')
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Cost Reduction During Training')
plt.show()
```

```python
x_new = 7
predicted_y = m * x_new + b

print(f'Prediction for X = {x_new}: {predicted_y:.2f}')
```
